# Understanding vaccination numbers
For the purposes of our research we will look at time periods before and after a critical majority of the at-risk population in Israel was fully vaccinated. For the purposes of our research we will define this as when 90% of people over 70 were vaccinated. We will use Israeli Ministry of health data on vaccination by age group.

```python
report_period = set(pd.date_range(
    vaccination['VaccinationDate'].min(),
    vaccination['VaccinationDate'].max(),
))

fill_back = {c: [] for c in vaccination.columns}

for category in vaccination['age_group'].unique():
    cat_records = vaccination[vaccination['age_group'] == category]
    cat_dates = set(pd.to_datetime(cat_records['VaccinationDate']))
    missing = report_period - cat_dates
    # print('Category: {} missing dates: {}'.format(category, missing))
    for m in missing:
        fill_back['VaccinationDate'].append(m.strftime('%Y-%m-%d'))
        fill_back['age_group'].append(category)
        fill_back['first_dose'].append('0')
        fill_back['second_dose'].append('0')

fill_back_df = pd.DataFrame(fill_back)
vaccination_patched = pd.concat((vaccination, fill_back_df))
vaccination = vaccination_patched
```

```python
# we will convert the date over to a date-time series and put it into a new column
vaccination['date_as_date'] = pd.to_datetime(vaccination['VaccinationDate'], format= '%Y-%m-%d')
# the other numeric data was provided as strings. we willconvert the strings to number
# and we will conservatively claim that when the number is recorded as '<15' it can
# be rounded down to zero. we do not want to overestimate the vaccinated population
# we will return to this assumption later and run the numbers if we round up to compare
vaccination['second_dose'] = vaccination['second_dose'].str.replace('<15', '0')
vaccination['first_dose'] = vaccination['first_dose'].str.replace('<15', '0')
vaccination['first_dose'] = pd.to_numeric(vaccination['first_dose'])
vaccination['second_dose'] = pd.to_numeric(vaccination['second_dose'])
```

## Examining trends

```python
s1 = skew(prevax_deaths_norm)
k1 = kurtosis(prevax_deaths_norm)

s2 = skew(postvax_deaths_norm)
k2 = kurtosis(postvax_deaths_norm)

print('prevax death skew: {}, kurtosis: {}'.format(s1, k1))
print('postvax death skew: {}, kurtosis: {}'.format(s2, k2))
```

So we can't actually say much about skew or kurtosis
We see that skew of deaths is positive in both cases, but more extreme before vaccination.
However, the way Python scipy calculates skew leaves much to be desired. After reading the documentation
carefully, for a sequence like this ...it's highly misleading...but just to be complete...

Kurtosis doesn't show an understandable trend, but that is probably to be expected, as we are not looking at a distribution around a mean, a gaussian or poisson-like distribution.  Let's look at cases:

```python
sc1 = skew(prevax_cases_norm)
kc1 = kurtosis(prevax_cases_norm)

sc2 = skew(postvax_cases_norm)
kc2 = kurtosis(postvax_cases_norm)

print('prevax cases skew: {}, kurtosis: {}'.format(sc1, kc1))
print('postvax cases skew: {}, kurtosis: {}'.format(sc2, kc2))
```

So now we see in cases the distribution's skews were both positive, and pretty close. The same with kurtosis, but these numbers have no place in the final analysis as they are highly misleading due to technicalities. 

### A more appropriate analysos that can be applied is to examine mean squared error over normalized sequences (we must normalize for differences in overall death rates in different periods if we want to compare distributions). We will look at similarity by mean squared error over the normalized sequences

```python
# mean squared error deaths
mean_square_error_deaths = ((np.float64(postvax_deaths_norm) - np.float64(prevax_deaths_norm))**2).sum()**0.5
max_possible_error_deaths = len(postvax_deaths_norm)
dissimilarity = mean_square_error_deaths / (max_possible_error_deaths / 100)
print('Two distributions for deaths are similar: {:.2f}% after normalization'.format(100 - dissimilarity))
```

```python
# mean squared error cases
mean_square_error_cases = ((np.float64(postvax_cases_norm) - np.float64(prevax_cases_norm))**2).sum()**0.5
max_possible_error_cases = len(postvax_cases_norm)
dissimilarity_cases = mean_square_error_cases / (max_possible_error_cases / 100)
print('Two distributions for cases are similar: {:.2f}% after normalization'.format(100 - dissimilarity_cases))
```
