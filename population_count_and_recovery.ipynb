```python
def aggregated_covid_recovery_transposed():
    all_recovered = pd.read_csv('D:/MAXCORlabwork/covid/most_current_csvs/corona_recovered_ver_00143.csv')
    daily_recovered = pd.read_csv('D:/MAXCORlabwork/covid/most_current_csvs/recovered_daily.csv')
    daily_recovered.columns.values[:] = 'date', 'recovered'
    daily_recovered.dropna(inplace=True)
    daily_recovered['recovered'] = daily_recovered['recovered'].astype('int32')
    all_recovered['date'] = ''
    recoverd_so_far = 0
    for i, row in daily_recovered.iterrows():
        rec = row['recovered']
        date = row['date']
        for i in range(rec):
            all_recovered.at[recoverd_so_far + i, 'date'] = date
        recoverd_so_far += rec
    all_recovered = all_recovered[all_recovered['date'] != '']
    all_recovered.drop(
        set(all_recovered.columns) - {'date', 'age_group'},
        axis=1,
        inplace=True,
    )
    age_groups = list(all_recovered['age_group'].unique())
    return all_recovered.groupby(
        ['date', 'age_group'],
    ).agg(count=('age_group', 'count')).reset_index()
```

```python
try_again = aggregated_covid_recovery_transposed()
```

```python
try_again = try_again.sort_values(by = 'date', ascending = False)
try_again = try_again.reset_index()
try_again['count'].max()
```

```python
try_again.to_csv('try1_daily_recovered_by_age.csv')
```
